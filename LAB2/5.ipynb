Wymagane funkcje:

1.	Wyświetlanie obrazu wczytanego z pliku o podanej nazwie.
    
2.	Sporządzanie wykresów zmian poziomu szarości wzdłuż wybranej linii poziomej lub pionowej o zadanej współrzędnej.

3.	Wybór podobrazu (prostokątnego obszaru) o podanych współrzędnych oraz jego zapis do pliku o zadanej nazwie.

```python
import cv2
import matplotlib.pyplot as plt
import tifffile

def display_image(image_path):
    img = tifffile.imread(image_path)
    plt.imshow(img, cmap='gray')
    plt.title('aerial_view.tif')
    plt.axis('off')
    plt.show()

def plot_intensity(image_path, x=None, y=None):
    img = tifffile.imread(image_path)
    if x is not None and y is None:
        intensity = img[:, x]  # Wybierz całą kolumnę o współrzędnej x
        plt.plot(intensity)
        plt.title('Zmiany poziomu szarości wzdłuż linii pionowej')
        plt.xlabel('Pozycja pionowa')
        plt.ylabel('Poziom szarości')
        plt.show()
    elif y is not None and x is None:
        intensity = img[y, :]  # Wybierz cały wiersz o współrzędnej y
        plt.plot(intensity)
        plt.title('Zmiany poziomu szarości wzdłuż linii poziomej')
        plt.xlabel('Pozycja pozioma')
        plt.ylabel('Poziom szarości')
        plt.show()
    else:
        print("Podaj tylko współrzędną x lub y.")

def select_and_save_subimage(image_path, x1, y1, x2, y2, output_path):
    img = tifffile.imread(image_path)
    subimage = img[y1:y2, x1:x2]
    tifffile.imwrite(output_path, subimage)
    print(f"Podobraz został zapisany jako: {output_path}")

# Przykładowe użycie:
image_path = 'pliki/aerial_view.tif'
display_image(image_path)
plot_intensity(image_path, x=100)
select_and_save_subimage(image_path, 50, 50, 150, 150, 'subimage.tif')
```

Skrypt umożliwia wczytywanie i wyświetlanie obrazów zapisanych w formacie TIFF. Funkcja plot_intensity umożliwia analizę zmian poziomu szarości wzdłuż wybranej linii (pionowej lub poziomej). Dzięki temu można łatwo zidentyfikować różnice w jasności w różnych częściach obrazu, co jest szczególnie przydatne w analizie obrazów o dużej rozdzielczości. Funkcja select_and_save_subimage pozwala na wybór prostokątnego obszaru obrazu i zapisanie go do nowego pliku. Jest to przydatne w sytuacjach, gdy konieczna jest analiza lub przechowywanie tylko wybranego fragmentu obrazu.
