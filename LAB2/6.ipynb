```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile


def multiply_image(image_path, c):
    img = tifffile.imread(image_path)
    transformed_img = np.clip(c * img, 0, 255).astype(np.uint8)
    plt.imshow(transformed_img, cmap='gray')
    plt.title('MnoÅ¼enie obrazu przez staÅ‚Ä…')
    plt.axis('off')
    plt.show()
    return transformed_img

def log_transform(image_path, c):
    img = tifffile.imread(image_path)
    transformed_img = c * np.log1p(img)
    transformed_img = np.clip(transformed_img, 0, 255).astype(np.uint8)
    plt.imshow(transformed_img, cmap='gray')
    plt.title('Transformacja logarytmiczna')
    plt.axis('off')
    plt.show()
    return transformed_img

def contrast_stretching(image_path, m, e):
    img = tifffile.imread(image_path)
    transformed_img = 1 / (1 + (m / (img + 1e-5))**e)
    transformed_img = np.clip(transformed_img * 255, 0, 255).astype(np.uint8)
    plt.imshow(transformed_img, cmap='gray')
    plt.title('Zmiana dynamiki skali szaroÅ›ci (kontrast)')
    plt.axis('off')
    plt.show()
    return transformed_img

image_path_multiply = 'pliki/chest-xray.tif'
image_path_transform = 'pliki/spectrum.tif'
image_path_contrast = 'pliki/chest-xray.tif'

transformed_image_1 = multiply_image(image_path_multiply, c=1.5)
transformed_image_2 = log_transform(image_path_transform, c=30)
transformed_image_3 = contrast_stretching(image_path_contrast, m=128, e=4)
```
