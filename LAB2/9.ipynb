```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tifffile

def average_filter(image_path, kernel_size):
    img = tifffile.imread(image_path)
    img = np.clip(img, 0, 255).astype(np.uint8)

    # Apply average (mean) filter
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    filtered_img = cv2.filter2D(img, -1, kernel)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(img, cmap='gray')
    axes[0].set_title('Obraz przed filtracjÄ…')
    axes[0].axis('off')

    axes[1].imshow(filtered_img, cmap='gray')
    axes[1].set_title('Obraz po uÅ¼yciu filtra uÅ›redniajÄ…cego')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()


    return filtered_img


image_path = 'pliki/cboard_pepper_only.tif'
average_filter(image_path, 3)
```

```python
def min_filter(image_path, kernel_size=3):
    img = tifffile.imread(image_path)
    img = np.clip(img, 0, 255).astype(np.uint8)

    # Apply minimum filter using erosion
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    filtered_img = cv2.erode(img, kernel)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(img, cmap='gray')
    axes[0].set_title('Obraz przed filtracjÄ…')
    axes[0].axis('off')

    axes[1].imshow(filtered_img, cmap='gray')
    axes[1].set_title('Obraz po uÅ¼yciu filtra minimum')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    return filtered_img


min_filter(image_path, 3)
```

```python
def max_filter(image_path, kernel_size=3):
    img = tifffile.imread(image_path)
    img = np.clip(img, 0, 255).astype(np.uint8)

    # Apply maximum filter using dilation
    kernel = np.ones((kernel_size, kernel_size), np.uint8)
    filtered_img = cv2.dilate(img, kernel)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(img, cmap='gray')
    axes[0].set_title('Obraz przed filtracjÄ…')
    axes[0].axis('off')

    axes[1].imshow(filtered_img, cmap='gray')
    axes[1].set_title('Obraz po uÅ¼yciu filtra maximum')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    return filtered_img


max_filter(image_path, 3)
```
