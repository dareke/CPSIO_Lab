```python
import numpy as np
import matplotlib.pyplot as plt
import tifffile
import cv2


def display_image(image_path):
    img = tifffile.imread(image_path)
    plt.imshow(img, cmap='gray')
    plt.title('Obraz')
    plt.axis('off')
    plt.show()

def equalize_histogram(image_path):
    img = tifffile.imread(image_path)

    # Przeskalowanie obrazu do zakresu [0, 255] jeÅ›li jest w innym zakresie
    img = np.clip(img, 0, 255).astype(np.uint8)

    # WyrÃ³wnanie histogramu
    equalized_img = cv2.equalizeHist(img)

    # WyÅ›wietlenie obrazÃ³w przed i po wyrÃ³wnaniu
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))

    axes[0, 0].imshow(img, cmap='gray')
    axes[0, 0].set_title('Obraz przed wyrÃ³wnaniem')
    axes[0, 0].axis('off')

    axes[0, 1].hist(img.flatten(), bins=256, range=[0, 256], color='gray')
    axes[0, 1].set_title('Histogram przed wyrÃ³wnaniem')

    axes[1, 0].imshow(equalized_img, cmap='gray')
    axes[1, 0].set_title('Obraz po wyrÃ³wnaniu')
    axes[1, 0].axis('off')

    axes[1, 1].hist(equalized_img.flatten(), bins=256, range=[0, 256], color='gray')
    axes[1, 1].set_title('Histogram po wyrÃ³wnaniu')

    plt.tight_layout()
    plt.show()

    return equalized_img


image_path = 'pliki/hidden-symbols.tif'
equalized_image = equalize_histogram(image_path)
```
