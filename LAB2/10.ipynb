```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tifffile

def avg_filter(image_path, size):
    img = tifffile.imread(image_path)
    img = np.clip(img, 0, 255).astype(np.uint8)

    # Apply average (mean) filter
    kernel = np.ones((size, size), np.float32) / (size * size)
    filtered_img = cv2.filter2D(img, -1, kernel)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(img, cmap='gray')
    axes[0].set_title('Obraz przed filtracjÄ…')
    axes[0].axis('off')

    axes[1].imshow(filtered_img, cmap='gray')
    axes[1].set_title('Obraz po uÅ¼yciu filtra uÅ›redniajÄ…cego')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    return filtered_img


image_path = 'pliki/characters_test_pattern.tif'
avg_filter(image_path, 3) #tu rozmiar wg wykladu 4. jest najlepszy okolo 3, im wiekszy tym wieksze rozmycie obrazu
```

```python
def gauss_filter(image_path, kernel_size, sigma):
    img = tifffile.imread(image_path)
    img = np.clip(img, 0, 255).astype(np.uint8)

    # Apply Gaussian filter
    filtered_img = cv2.GaussianBlur(img, (kernel_size, kernel_size), sigma)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(img, cmap='gray')
    axes[0].set_title('Obraz przed filtracjÄ…')
    axes[0].axis('off')

    axes[1].imshow(filtered_img, cmap='gray')
    axes[1].set_title('Obraz po uÅ¼yciu filtra gaussowskiego')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    return filtered_img


gauss_filter(image_path, 5, 7) #sigma=7 z wykladu, rozmiar tak samo jak w filtrze wyzej
```
